```python
# Churn Distribution

print(df['Churn'].value_counts())
```

```python
# Churn Percentage

churn_rate = df['Churn'].value_counts(normalize=True) * 100
print(churn_rate)
```

```python
import matplotlib.pyplot as plt

df['Churn'].value_counts().plot(kind='bar')

plt.title('Customer Churn Distribution')
plt.xlabel('Churn')
plt.ylabel('Number of Customers')

plt.show()
```

```python
gender_churn = pd.crosstab(df['gender'], df['Churn'])

print(gender_churn)
```

```python
gender_churn.plot(kind='bar')

plt.title('Churn by Gender')
plt.xlabel('Gender')
plt.ylabel('Customers')

plt.show()
```

```python
contract_churn = pd.crosstab(df['Contract'], df['Churn'])

print(contract_churn)
```

```python
contract_churn.plot(kind='bar')

plt.title('Churn by Contract Type')
plt.xlabel('Contract')
plt.ylabel('Customers')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

df.boxplot(column='MonthlyCharges', by='Churn')

plt.title('Monthly Charges by Churn')
plt.suptitle('')
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')

plt.show()
```

```python
plt.figure(figsize=(8,5))

df.boxplot(column='tenure', by='Churn')

plt.title('Tenure by Churn')
plt.suptitle('')
plt.xlabel('Churn')
plt.ylabel('Tenure (Months)')

plt.show()
```

```python
internet_churn = pd.crosstab(df['InternetService'], df['Churn'])

print(internet_churn)
```

```python
internet_churn.plot(kind='bar')

plt.title('Internet Service vs Churn')
plt.xlabel('Internet Service')
plt.ylabel('Customers')

plt.show()
```

```python
payment_churn = pd.crosstab(df['PaymentMethod'], df['Churn'])

print(payment_churn)
```

```python
payment_churn.plot(kind='bar', figsize=(10,5))

plt.title('Payment Method vs Churn')
plt.xlabel('Payment Method')
plt.ylabel('Customers')

plt.show()
```

# Business Insights

1. Customers on month-to-month contracts have the highest churn rate.

2. Customers with shorter tenure are more likely to leave the company.

3. Customers paying higher monthly charges show a greater tendency to churn.

4. Fiber optic internet users experience higher churn compared to DSL users.

5. Customers using Electronic Check as their payment method have relatively higher churn rates.

6. Long-term contract customers demonstrate better retention and lower churn.
